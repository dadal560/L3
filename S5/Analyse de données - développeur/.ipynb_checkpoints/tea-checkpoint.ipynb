```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.matlib import repmat
import warnings
from numpy.linalg import eigh as eigen
warnings.filterwarnings('ignore')
```

## Copiez-collez vos fonctions Python de standardisation (TP1) et d'analyse en composantes principales (TP2), puis testez-les sur les données *terminale*.

```python
def standardize(data, reduire = False):
    xbar = np.mean(data,axis = 0)
    Xp = data - xbar
    n,p = data.shape
    D = np.eye(n) / n
    V = Xp.T @ D @ Xp

    if (reduire == False) :
        D = np.eye(n) / n
        return (Xp, V)
    else : 
        stds = np.diag(V)**.5
        Xs = Xp / stds
        return (Xs, Xs.T@D@Xs)


def acpcanon(processed_data, processed_cov):
    L, U = eigen(processed_cov)
    L = np.sort(L)[::-1]
    U = U[:, L.argsort()]
    Cp = processed_data @ U

    return L,U,Cp
```

```python
dataf = pd.read_excel("Terminale.xls",index_col=0)
row_label = dataf.index.values 
col_label = dataf.columns.values
X = dataf[col_label].values
n, p = X.shape
```

```python
Xp, V = standardize(X)
L,U,C = acpcanon(Xp,V)
```

# 1) Quel taux de compression de données est réalisé avec une ACP ?
***
## Pour reconstruire le tableau centré $X'$ à partir de $q$ composantes, il faut les avoir conservées, ainsi que les vecteurs directeurs correspondants, à savoir : $C'_{(1:q)}$ et les $U_{(1:q)}$. Il ne faut pas non oublier de décentrer ; pour cela, il faut aussi conserver $\overline{x}$. C'est pourquoi le taux de compression peut être supérieur à 1 si $q$ n'est pas suffisament petit devant $p$ ! 

***
## A l'aide d'une boucle sur les composantes, calculez le taux de compression en fonction du nombre *q* de composantes retenues.

```python
taux = []
for q in range(p):
    taux.append(1-(n*q+p*q+p)/ n/p)
    
```

## Affichez la courbe du *taux* en fonction de *q*. Commentez brièvement le résultat.

```python
plt.plot(number,taux)
plt.plot(number,(n-1)/n-(n+p)/n/p*number)
plt.grid(True)
```

## Eh oui ! le taux de compression est une fonction affine de $q$.
## Petit exercice facile d'arithmétique : laquelle ?

## Le taux de compression peut donc être supérieur à 1 si $q$ n'est pas suffisamment petit devant $p$ ! 

## Petit exercice facile d'arithmétique : trouvez, la borne supérieure sur $q$ pour qu'il y ait vraiment compression de données ($t(q)<1$). Cett borne $q_{max}$  $n$ et $p$.

```python
qmax = p*(n-1) // (n+p)
qmax
```

```python
tmax = (n*qmax + p*qmax + p) / n / p
t = 1/tmax
```
